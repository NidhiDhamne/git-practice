```python
import spacy
from spacy import displacy
from spacy import tokenizer
nlp = spacy.load('en_core_web_sm')
```

```python
text =("Koalas, native to Australia, are renowned for their cuddly appearance with soft grayish-brown fur, large fluffy ears, and black noses. These herbivores primarily feed on tough eucalyptus leaves, relying on their specialized digestive systems to process them. Koalas are famous for their leisurely lifestyle, sleeping up to 20 hours a day in trees, where they are skilled climbers. While often seen as solitary animals, they communicate through vocalizations and scent markings. Unique among marsupials, baby koalas, or joeys, are born undeveloped and continue growing in their mother's pouch. Unfortunately, they face threats like habitat loss, making conservation efforts crucial for their survival.")
```

```python
doc = nlp(text)

sentences = list(doc.sents)
print(sentences)

# tokenization
for token in doc:
    print(token.text)
    
# print entities
ents = [(e.text, e.start_char, e.end_char, e.label_) for e in doc.ents]
print(ents)

# now we use displaycy function on doc2
displacy.render(doc, style='ent', jupyter=True)
```
